```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../datasets/social_media_data.csv")

df.head()
```

```python
df.info()
df.isnull().sum()
```

```python
platform_likes = df.groupby("Platform")["Likes"].sum()

platform_likes.plot(kind="bar")
plt.title("Total Likes by Platform")
plt.xlabel("Platform")
plt.ylabel("Likes")
plt.tight_layout()
plt.show()
```

```python
sns.barplot(x="PostType", y="Likes", data=df)
plt.title("Average Likes by Post Type")
plt.tight_layout()
plt.show()
```

```python
sns.scatterplot(x="Comments", y="Likes", data=df)
plt.title("Likes vs Comments")
plt.tight_layout()
plt.show()
```

```python
df[["Likes","Comments","Shares","Reach"]].corr()
```
