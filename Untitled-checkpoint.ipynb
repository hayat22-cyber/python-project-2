```python
import pandas as pd

df = pd.read_csv(
    r"C:\Users\hello\OneDrive\Documents\project4_finance_analysis\datasets\finance_data.csv"
)

print(df.head())
print(df.info())
```

```python
# Convert Date to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Fill missing values if any
df.ffill(inplace=True)

# Check missing values
print(df.isnull().sum())
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,4))
for stock in df['Stock'].unique():
    stock_data = df[df['Stock'] == stock]
    plt.plot(stock_data['Date'], stock_data['Close'], label=stock)

plt.title("Stock Closing Prices Over Time")
plt.xlabel("Date")
plt.ylabel("Close Price")
plt.legend()

plt.savefig(
    r"C:\Users\hello\OneDrive\Documents\project4_finance_analysis\visualizations\stock_trends.png"
)
plt.show()
```

```python
df['Return'] = df.groupby('Stock')['Close'].pct_change() * 100
print(df)
```

```python
import seaborn as sns

plt.figure(figsize=(6,4))
sns.histplot(df['Return'].dropna(), bins=10, color='green')
plt.title("Distribution of Daily Returns")
plt.xlabel("Return (%)")
plt.ylabel("Frequency")

plt.savefig(
    r"C:\Users\hello\OneDrive\Documents\project4_finance_analysis\visualizations\daily_returns.png"
)
plt.show()
```

```python
df.to_csv(
    r"C:\Users\hello\OneDrive\Documents\project4_finance_analysis\reports\finance_cleaned.csv",
    index=False
)

print("Cleaned CSV saved in reports folder")
```
