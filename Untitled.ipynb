```python
import pandas as pd

df = pd.read_csv(
    r"C:\Users\hello\OneDrive\Documents\project3_weather_analysis\datasets\weather_data.csv"
)

print(df.head())
```

```python
# Convert Date column to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Fill missing values (if any)
df.fillna(method='ffill', inplace=True)

# Check again
print(df.isnull().sum())
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,4))
plt.plot(df['Date'], df['Temperature'])
plt.title("Temperature Trend Over Time")
plt.xlabel("Date")
plt.ylabel("Temperature")

plt.savefig(
    r"C:\Users\hello\OneDrive\Documents\project3_weather_analysis\visualizations\temperature_trend.png"
)

plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(6,4))
sns.histplot(df['Rainfall'], bins=20, color='blue')
plt.title("Rainfall Distribution")
plt.xlabel("Rainfall")
plt.ylabel("Frequency")

plt.savefig(
    r"C:\Users\hello\OneDrive\Documents\project3_weather_analysis\visualizations\rainfall_distribution.png"
)
plt.show()
```
